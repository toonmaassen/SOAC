```python
import numpy as np
import matplotlib.pyplot as plt
import math


# Calculating the 1-year CO2-e of methane.
# Decay model from Muller & Muller (2017): https://www.scitechnol.com/peer-review/fugitive-methane-and-the-role-of-atmospheric-halflife-u53c.php?article_id=6097
Nyears=400
co2decaylist=np.zeros((Nyears))
ch4decaylist=np.zeros((Nyears))
for t in range(Nyears):
    co2decaylist[t]=0.217 + 0.259*np.exp(-0.1*t/172.9)+ 0.338*np.exp(-0.1*t/18.51)+ 0.186*np.exp(-0.1*t/1.186) #decay of CO2 (Muller & Muller, 2017)
    ch4decaylist[t]=0.5**(0.1*t/8.6) #decay of methane  (Muller & Muller, 2017) 
plt.plot(np.arange(0,Nyears/10,0.1),co2decaylist,label='CO$_2$')
plt.fill_between(np.arange(0,20,0.1),co2decaylist[0:200], color='lightsteelblue')
plt.plot(np.arange(0,Nyears/10,0.1),ch4decaylist,label='CH$_4$')
plt.fill_between(np.arange(0,20,0.1),ch4decaylist[0:200], color='bisque')
plt.ylim(0,1)
plt.ylabel('Fraction of molecules remaining')
plt.xlabel('Time (years)')
#plt.title('Decay of CO$_2$ and CH$_4$')
plt.legend()
plt.show()

#20 year CO2-e times the ratio of the area under the decay graphs over 20 years.
methaneco2e1yr=80.8*np.sum(co2decaylist[0:200])/np.sum(ch4decaylist[0:200]) 
print('The 1-year CO2-e of methane is '+str(round(methaneco2e1yr,2))+'.')
```
